```python
n=int(input())
tong=0
for i in range(len(str(n))):
    if int(str(n)[i])%3==0:
        tong+=int(str(n)[i])
print(tong)
```

```python
n=int(input())
lst=list(map(int,input().split()))
flag=True
for i in range(1,n-1):
    if lst[i+1]-lst[i]!=lst[i]-lst[i-1]:
        flag=False
if flag:
    print("YES")
else:
    print("NO")
```

```python
def snt(n):
    if n<=1:
        return False
    elif n%2==0 and n!=2:
        return False
    for i in range(3,int(n**0.5)+1,2):
        if n%i==0:
            return False
    return True
n=int(input())
lst=list(map(int,input().split()))
dem=0
for i in lst:
    if snt(i):
        dem+=1
print(dem)
```

```python
n=int(input())
lst=list(map(int,input().split()))
tongchan=0
tongle=0
for i in lst:
    if i%2==0:
        tongchan+=i
    else:
        tongle+=i
print(tongchan,tongle)
```

```python
n=input()
nr=int(n[::-1])
for i in range(len(str(nr))):
    if int(str(nr)[i])!=0:
        break
    else:
        nr.remove((nr[i]))
print(int(nr))
```
